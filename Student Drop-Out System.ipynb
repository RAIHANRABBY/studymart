```python
f_income=[]

for i in df['Father Occupation']:
    if i == 'Immigrant':
        f_income.append('Remittance')
    elif i=='Freelancer':
        f_income.append('Remittance')
    elif i=='Private sector':
        f_income.append('Casual Wage')
    elif i =='Public sector':
        f_income.append('Formal Wage')
    elif i == 'Unemployed':
        f_income.append('Pansion')
    elif i == 'Worker':
        f_income.append('Casual Wage')
    elif i =='Agriculture':
        f_income.append('Farming')
    else:
        f_income.append('Own business')
f_income
        
```

```python
df['Family income source']=f_income
df.head()
```
