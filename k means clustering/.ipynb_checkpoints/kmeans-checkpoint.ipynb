```python
import seaborn as sns
```

```python
sns.pairplot(df[['gender','income','score']])
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
```

```python
df['gender']=encode.fit_transform(df['gender'])
```

```python
sns.pairplot(df[['Age','gender','score','income']])
```

```python
from sklearn.cluster import KMeans
```

```python
kmean=KMeans(n_clusters=5)
```

```python
karmy=kmean.fit(df[['score','income']])
```

```python
karmy.cluster_centers_
```

```python
df['income_cluster']=kmean.labels_
```

```python
df.head()
```

```python
sns.scatterplot(x='income',y='score',hue='income_cluster',data=df)
```

```python
df['income_cluster'].value_counts()
```
